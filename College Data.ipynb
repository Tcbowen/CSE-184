```python
Location = r'C:\Users\tomb3\Desktop\S-F data.csv'
df = pd.read_csv(Location)
print(df.columns)
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.scatter(df['S-F ratio'], df['Grad rate '], alpha = .2, color = 'r')
ax2.scatter(df['S-F ratio'], df['retention rate'], alpha = .2, color = 'r')
ticks = ax1.set_xticks([0, 5, 10, 15, 20, 25])
ticks = ax2.set_xticks([0, 5, 10, 15, 20, 25])
SF_data = df[['Insitution Name ', 'S-F ratio','Grad rate ', 'retention rate']]
SF_data
```

```python
merged_data = pd.merge(SF_data, Sati_data, left_on = 'Insitution Name ', right_on='INSTNM')
```

```python
merged_data = merged_data.drop(['INSTNM'], axis =1)
merged_data
```
