```python
import pandas as pd
import matplotlib.pyplot as plt
Location = r'C:\Users\tomb3\Desktop\UCSC files.csv'
df = pd.read_csv(Location)
Quat = {"FALL": .99, "WINTER":.66,"SPRING":.33}
def term_to_num(x):
    term = x.split(' ')
    return Quat[term[0]]+int(term[1]) 
df['term_num'] = df['term'].apply(term_to_num)
df
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1, 1, 1)
ax1.scatter(df['term_num'], df['num_people_enrolled'], alpha = .3, color = 'r')
ticks = ax1.set_yticks([0, 50, 100, 150, 200, 250,300, 350, 400, 450, 500])
plt.xticks(rotation=45)
```

```python
df_means = df.groupby('term').mean()
df_means
df_means = df_means.sort_values(by = 'term_num')
df_means
df_means['open'] = df_means['total_class_size'] - df_means['num_people_enrolled']
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(df_means['num_people_enrolled'], alpha = 1, color = 'r')
ticks = ax1.set_yticks([0, 20, 40, 60, 80, 100,120, 140])
plt.xticks(rotation=45)
```

```python
## totals 
df_totals = df.groupby('term').sum()
df_totals = df_totals.sort_values(by = 'term_num')
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(df_totals['num_people_enrolled'], alpha = 1, color = 'r')
ticks = ax1.set_yticks([0, 1000, 2000, 3000, 4000, 5000, 6000, 7000])
plt.xticks(rotation=45)
```
