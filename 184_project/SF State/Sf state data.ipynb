```python
import pandas as pd
import matplotlib.pyplot as plt
Location = r'C:\Users\tomb3\Desktop\SF_State_Data.csv'
df = pd.read_csv(Location)
Quat = {"Fall": .8, "Spring":.4}
def term_to_num(x):
    term = x.split(' ')
    return Quat[term[0]]+int(term[1]) 
df['term_num'] = df['term'].apply(term_to_num)
def course_to_num(x):
    try:
        return int(x.split(" ")[1].split('[')[0])
    except:
        return 0
df['Course_num'] = df['Course_Name'].apply(course_to_num)
undergrad = df.query('Course_num<700')
```

```python
df_means = df.groupby('term').mean()
df_means
df_means = df_means.sort_values(by = 'term_num')
df_means
df_means['open'] = df_means['total_Seats'] - df_means['enrolled']
```

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot(df_means['enrolled'], alpha = 1, color = 'r')
ticks = ax1.set_yticks([0, 5, 10, 15, 20, 25])
plt.xticks(rotation=45)
```
